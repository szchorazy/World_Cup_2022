```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/Admin/Desktop/Fifa_world_cup_matches.csv', engine='python')
```

```python
df.head()
```

```python
df['possession team1'] = df['possession team1'].str.replace('%','').astype(int)
df['possession team2'] = df['possession team2'].str.replace('%','').astype(int)
```

```python
df.head()
```

```python
df.columns
```

```python
def pie_chart(team1,team2):
    win_1 = 0
    lose_1 = 0
    draw_1 = 0
    for i in range(len(df)):
        if df[team1][i]>df[team2][i] and df["number of goals team1"][i]>df['number of goals team2'][i]:
            win_1+= 1
        elif df[team1][i]<df[team2][i] and df["number of goals team1"][i]<df['number of goals team2'][i]:
            win_1+=1
        elif df[team1][i]>df[team2][i] and df["number of goals team1"][i]<df['number of goals team2'][i]:
            lose_1+=1
        elif df[team1][i]<df[team2][i] and df["number of goals team1"][i]>df['number of goals team2'][i]:
            lose_1+=1
        elif df[team1][i]>df[team2][i] and df["number of goals team1"][i]==df['number of goals team2'][i]:
            draw_1+=1
        elif df[team1][i]>df[team2][i] and df["number of goals team1"][i]==df['number of goals team2'][i]:
            draw_1+=1
    return win_1,lose_1,draw_1

def display_data(data,title):
    print (title)
    print("Win% =", str((data[0]/sum(data))*100)+"%")
    print("Lose% =",str((data[1]/sum(data))*100)+"%")
    print("Draw% =",str((data[2]/sum(data))*100)+"%","\n")
```

```python
win_pose, lose_pose,draw_pose= pie_chart('possession team1','possession team2')
win_dpa,lose_dpa,draw_dpa = pie_chart('defensive pressures applied team1','defensive pressures applied team2')
win_ft , lose_ft , draw_ft = pie_chart('forced turnovers team1','forced turnovers team2')
win_ta,lose_ta, draw_ta = pie_chart('total attempts team1','total attempts team2')
win_re, lose_re, draw_re = pie_chart('receptions between midfield and defensive lines team1','receptions between midfield and defensive lines team2')
win_br, lose_br, draw_br = pie_chart('passes completed team1','passes completed team2')

data_list = [[win_pose, lose_pose,draw_pose],[win_dpa,lose_dpa,draw_dpa],[win_ft , lose_ft , draw_ft],[win_ta,lose_ta, draw_ta],[win_re, lose_re, draw_re],[win_br, lose_br, draw_br]]
data_title = ["Posession","Defensive Pressures","Force Turnovers","Total Attempts",'Reception','Break']
new_title = ["Posessing ball","applying defensive pressure more on your oppponents","force turnover","high number of attempts","Reception","Break"]
```

```python
fig,axs = plt.subplots(3 , 2, figsize=(20,12))
axs[0,0].pie([win_pose,lose_pose,draw_pose],labels =["Win","Loss" , "Draw"],autopct='%1.1f%%')
axs[0,0].set_title("Does higher ball possesion help you win the game?")
axs[0,1].pie([win_dpa,lose_dpa,draw_dpa],labels =["Win","Loss" , "Draw"],autopct='%1.1f%%')
axs[0,1].set_title("Does applying defensive pressure more on your oppponents help you win the game?")
axs[1,0].pie([win_ft,lose_ft,draw_ft],labels = ["Win","Loss" , "Draw"],autopct='%1.1f%%')
axs[1,0].set_title("Does force turnover help you win the game?")
axs[1,1].pie([win_ta,lose_ta,draw_ta],labels = ["Win","Loss" , "Draw"],autopct='%1.1f%%')
axs[1,1].set_title("Does higher number of attempts help you win the game?")
axs[2,1].pie([win_re, lose_re, draw_re],labels = ["Win","Loss" , "Draw"],autopct='%1.1f%%')
axs[2,1].set_title("Does higher number of receptions help win the game?")
axs[2,0].pie([win_br, lose_br, draw_br ],labels = ["Win","Loss" , "Draw"],autopct='%1.1f%%')
axs[2,0].set_title("Does higher number of completed passes help win the game? ")
```
